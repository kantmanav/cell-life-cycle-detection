```python
import pandas as pd

app_img_type = "Resize"
embedding_layer = "LSTM"
app = app_img_type.lower()
embedding = embedding_layer.lower()

embeddings_df = pd.read_csv('/data/large_files/embeddings/' + str(app) + '_' + str(embedding) + '_embeddings_pca.csv', index_col=[0])
```

```python
embeddings_df[33:37]
```

```python
time_sects = []
for frame in embeddings_df.loc[:, 'Frame']:
    if frame != 70:
        time_sects.append(frame // 7)
    else:
        time_sects.append(9)
embeddings_df['Section 1'] = time_sects

time_sects = []
for frame in embeddings_df.loc[:, 'Frame']:
    if frame != 70:
        time_sects.append(frame // 14)
    else:
        time_sects.append(4)
embeddings_df['Section 2'] = time_sects
```

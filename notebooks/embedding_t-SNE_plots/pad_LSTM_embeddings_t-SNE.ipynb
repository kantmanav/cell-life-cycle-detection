```python
# No initialization, PCA >50%, PCA >90%
# Default perplexity, high perplexity

import pandas as pd
embeddings_df = pd.read_csv('/data/large_files/embeddings/pad_lstm_embeddings.csv', index_col=[0])
```

```python
pd.set_option("display.max_columns", None)
```

```python
embeddings_df[:40]
```

```python
from sklearn.manifold import TSNE
tsne_def = TSNE(n_components=2)
```

```python
def project_embeddings_tsne(embeddings_df, tsne):
    neighbors_embeddings = tsne.fit_transform(embeddings_df)
    neighbors_df = pd.DataFrame(data = neighbors_embeddings, columns = ['component 1', 'component 2'])
    
    final_df = pd.concat([neighbors_df, embeddings_df[['Frame']]], axis = 1)
                          
    return final_df
```

```python
# No initialization, default perplexity

embeddings_tsne_df = project_embeddings_tsne(embeddings_df, tsne_def)
```

```python
import matplotlib.pyplot as plt

colors = []
for time in embeddings_tsne_df['Frame']:
    time_color = 0.7 * (time / 70)
    colors.append((time_color, time_color, time_color))

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Component 1', fontsize = 15)
ax.set_ylabel('Component 2', fontsize = 15)
ax.set_title('3T3 Cell Image Embeddings t-SNE Plot', fontsize = 20)
ax.scatter(embeddings_tsne_df['component 1'], embeddings_tsne_df['component 2'], c = colors, s = 50)
ax.legend()
ax.grid()
plt.savefig('../embedding_graphs/pad_lstm_embeddings_t_SNE/pad_lstm_embeddings_t_SNE_no_init_default_perp.jpeg')
```

```python
cols = ['Component 1', 'Component 2', 'Frame']
embeddings_tsne_df.to_csv('/data/large_files/embeddings/pad_lstm_embeddings_t_SNE_no_init_default_perp.csv', header=cols)
```

```python
# No initialization, high perplexity
tsne_high = TSNE(n_components=2, perplexity=816.56)
```

```python
%%time
embeddings_tsne_df = project_embeddings_tsne(embeddings_df, tsne_high)
```

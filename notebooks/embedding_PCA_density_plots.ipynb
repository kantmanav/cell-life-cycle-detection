```python
"""
0 - 6
7 - 13
14 - 20
21 - 27
28 - 34
35 - 41
42 - 48
49 - 55
56 - 62
63 - 70
"""
embeddings_df = pd.read_csv('/data/large_files/embeddings/resize_dense_embeddings_pca.csv', index_col=[0])

time_sects = []
for frame in embeddings_df.loc[:, 'Frame']:
    if frame != 70:
        time_sects.append(frame // 7)
    else:
        time_sects.append(9)
embeddings_df['Section'] = time_sects
```

```python
embeddings_df[33:37]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.displot(embeddings_df, x="component 1", hue="Section", binwidth=3, stat="density", common_norm=False, element="step").set(title="Component 1 Density Plot")
plt.savefig('embedding_graphs/resize_dense_embeddings_pca/density_plots/resize_dense_comp_1_sects_10_wide.jpg')
```

```python
sns.displot(embeddings_df, x="component 1", hue="Section", stat="density").set(title="Component 1 Density Plot")
plt.savefig('embedding_graphs/resize_dense_embeddings_pca/density_plots/resize_dense_comp_1_sects_10.jpg')
```

```python
sns.displot(embeddings_df, x="component 1", hue="Section", stat="density").set(title="Component 1 Density Plot")
plt.xlim(-20, 45)
plt.ylim(0, 0.03)
plt.savefig('embedding_graphs/resize_dense_embeddings_pca/density_plots/resize_dense_comp_1_sects_10_zoom.jpg')
```
